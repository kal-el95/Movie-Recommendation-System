# Creating a content based movie recommendation system 

```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
credits=pd.read_csv("tmdb_5000_credits.csv")
```

# Columns to consider:
    * movie_id
    * title
    * overview
    * genres
    * keywords
    * cast
    * crew

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.isna().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0]
```

```python
movies.iloc[0].genres
```

* Only considering the value for name key in genre and keywords

```python
import ast
```

```python
def convert(obj):
    lis=[]
    for i in ast.literal_eval(obj):
        lis.append(i['name'])
    return lis
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

* Consider only the first 3 actors from the cast and only the director from the crew

```python
def convert2(obj):
    lis = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 3:
            lis.append(i['name'])
        counter+=1
    return lis 
```

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
movies.head()
```

```python
def convert3(obj):
    lis = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            lis.append(i['name'])
    return lis 
```

```python
movies['crew']=movies['crew'].apply(convert3)
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
def collapse(obj):
    lis = []
    for i in obj:
        lis.append(i.replace(" ",""))
    return lis
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

* Concatinate all the columns into one

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new['tags'][0]
```

```python
new['tags']=new['tags'].apply(lambda x:x.lower())
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    lis=[]
    for i in text.split():
        lis.append(ps.stem(i))
    return " ".join(lis)
```

```python
new['tags']=new['tags'].apply(stem)
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

* Use cosine distance because in higher dimensionity euclidean distance is not reliable

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = similarity[index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
    li=[]
    for i in movies_list:
        li.append(new.iloc[i[0]].title)
    return li
```

```python
li=recommend('Batman Begins')
li
```

```python
import gradio as gr
```

```python
movie_titles = new['title'].tolist()
def get_recommendations(movie_name):
    recommendations = recommend(movie_name)
    recommendations_str = "".join([f"{i+1}. {movie}\n" for i, movie in enumerate(recommendations)])
    return recommendations_str
```

```python
iface = gr.Interface(
    fn=get_recommendations,
    inputs=gr.Dropdown(choices=movie_titles, label="Select a Movie"), 
    outputs=gr.Textbox(label="Recommended Movies"),
    title="Movie Recommendation System",
    description="Type in a movie name to get 10 similar movie recommendations."
)
```

```python
iface.launch()
```
